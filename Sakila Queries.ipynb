```python
engine = create_engine(connection)
engine
```

```python
# Query 1
query1 = """
    SELECT c.first_name, c.last_name, c.email, a.address, ci.city
    FROM customer c
    JOIN address a ON c.address_id = a.address_id
    JOIN city ci ON a.city_id = ci.city_id
    WHERE ci.city_id = 312;
"""
result1 = pd.read_sql_query(query1, engine)
```

```python
# Quer y 2
query2 = '''
    SELECT f.title, f.description, f.release_year, f.rating, f.special_features
    FROM film f
    JOIN film_category fc ON f.film_id = fc.film_id
    JOIN category c ON fc.category_id = c.category_id
    WHERE c.name = 'Comedy';
'''
result2 = pd.read_sql_query(query2, engine)
```

```python
# Query 3
query3 = '''
    SELECT a.last_name, f.title, f.release_year
    FROM actor a
    JOIN film_actor fa ON a.actor_id = fa.actor_id
    JOIN film f ON fa.film_id = f.film_id
    WHERE a.first_name = 'Johnny' AND a.last_name = 'Lollobrigida';
'''
result3 = pd.read_sql_query(query3, engine)
```

```python
# Query 4
query4 = '''
    SELECT a.first_name, a.last_name
    FROM actor a
    JOIN film_actor fa ON a.actor_id = fa.actor_id
    JOIN film f ON fa.film_id = f.film_id
    WHERE f.title = 'Bingo Talented';
'''
result4 = pd.read_sql_query(query4, engine)
```

```python
# Query 5
query5 = '''
    SELECT p.customer_id, p.amount
    FROM payment p
    WHERE p.amount > (SELECT AVG(amount) * 2 FROM payment);
'''
result5 = pd.read_sql_query(query5, engine)
```

```python
# Query 6
query6 = '''
    SELECT c.first_name, c.last_name, COUNT(p.payment_id) as num_payments
    FROM customer c
    JOIN payment p ON c.customer_id = p.customer_id
    GROUP BY c.customer_id
    ORDER BY num_payments DESC
    LIMIT 5;
'''
result6 = pd.read_sql_query(query6, engine)
```
